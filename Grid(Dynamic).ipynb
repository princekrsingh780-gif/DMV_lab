```python
# Draw Dynamic Grid (True & False)

import matplotlib.pyplot as plt

# Taking input from user
x = list(map(int, input("Enter X values (space-separated): ").split()))
y = list(map(int, input("Enter Y values (space-separated): ").split()))

# Check if lengths match
if len(x) != len(y):
    print("Error: X & Y must have same number of values")
else:
    plt.plot(x, y, marker='o')
    plt.grid(True)

    plt.xlabel("X values")
    plt.ylabel("Y values")
    plt.title("Graph with Grid")

    plt.show()
```
