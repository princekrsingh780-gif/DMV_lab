```python

# STEP 1: Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")


# STEP 2: Load Dataset

df = pd.read_csv("Titanic-Dataset.csv")   # make sure name matches your file


# STEP 3: Slice (30 rows × 10 cols)

df_small = df.iloc[:30, :10]

print("Shape:", df_small.shape)
display(df_small.head())
```

```python

# STEP 4: Basic Info

df_small.info()
display(df_small.describe())
```

```python

# STEP 5: Missing Values
print(df_small.isnull().sum())

plt.figure(figsize=(8,5))
sns.heatmap(df_small.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python

# STEP 6: Outliers (Boxplot)

num_cols = df_small.select_dtypes(include=['int64','float64']).columns

for col in num_cols:
    plt.figure()
    sns.boxplot(x=df_small[col])
    plt.title(f"Outliers in {col}")
    plt.show()
```

```python
#DATA VISUALIZATION

plt.figure()
plt.plot(df_small['PassengerId'], df_small['Fare'])
plt.title("Fare Trend")
plt.xlabel("PassengerId")
plt.ylabel("Fare")
plt.show()
```

```python
survival_counts = df_small['Survived'].value_counts()

plt.figure()
plt.bar(survival_counts.index, survival_counts.values)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
gender_counts = df_small['Sex'].value_counts()

plt.figure()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure()
plt.step(df_small['PassengerId'], df_small['Fare'], where='mid')
plt.title("Stair Plot of Fare")
plt.xlabel("PassengerId")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure()
df_small['Age'].hist(bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
